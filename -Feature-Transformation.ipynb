## <font color=Pink>Types Of Transformation :</font>

#### `1. Normalization And Standardization :`
Also known as Z-score normalization, this technique scales the features to have a mean of 0 and a standard deviation of 1. This is done by subtracting the mean and dividing by the standard deviation of each feature. <br>z=(x - x_mean)/std <br> - Standardization comes into picture when features of input data set have large differences between their ranges, or simply when they are measured in different measurement units (e.g., Pounds, Meters, Miles … etc).

#### `2. Scaling to Minimum And Maximum values : (OR : Max Absolute Scaling)`
This technique scales the features to a fixed range of values, typically between 0 and 1. This is done by subtracting the minimum value and dividing by the range of each feature. <br> X_scaled = [ X - X.min / (X.max - X.min) ] <br>
Max Absolute Scaling : This technique scales the features to a range of -1 to 1, by dividing each value by the maximum absolute value of that feature.

#### `3. Robust Scaling : Scaling To Median And Quantiles :`
This technique is similar to standardization but uses the median and interquartile range instead of the mean and standard deviation. This makes it more robust to outliers. X_scaled = (X - X.median) / IQR

#### `9. Box Cox Transformation :`

T(Y)=(Y exp(λ)−1)/λ  ,  Y is the response variable and λ is the transformation parameter.

The Box-Cox transformation is a power transformation technique that is used to normalize the distribution of a feature. 

Here's how the Box-Cox transformation process works:

Determine the range of lambda values to test. Lambda (λ) is a parameter that controls the degree of transformation. Typically, a range of values is tested, such as λ = -5, -4, -3, ..., 3, 4, 5.

For each lambda value, compute the transformed feature values using the Box-Cox formula:

y = [(x^λ) - 1] / λ for λ not equal to 0
log(x) for λ = 0

where x is the original feature value and y is the transformed feature value.

If λ is not zero, then the transformed values are scaled to have a mean of zero and a standard deviation of one.

The optimal value of the λ parameter can be found by maximizing a likelihood function or by cross-validation. It's important to note that the Box-Cox transformation can only be applied to features with strictly positive values.

The Box-Cox transformation can be useful in situations where the distribution of the feature is highly skewed or has outliers. By transforming the feature values, the Box-Cox transformation can help to normalize the distribution, making it more suitable for modeling purposes.



#### `10. Yeo-Johnson transformation :` 

The Yeo-Johnson transformation is a power transformation technique that is used to normalize the distribution of a feature. It is a modification of the Box-Cox transformation that can handle both positive and negative values.

Here's how the Yeo-Johnson transformation process works:

Add a constant to the feature values to make them all positive. This is done by adding a constant value that is greater than or equal to the absolute value of the smallest negative value in the feature.

Apply the transformation function to each feature value. The transformation function is defined as:

y = ((1 + x)^λ - 1) / λ for x ≥ 0
    -((1 - x)^(-λ) - 1) / λ for x < 0

where x is the original feature value, y is the transformed feature value, and λ is a parameter that determines the degree of transformation.

If λ is not zero, then the transformed values are scaled to have a mean of zero and a standard deviation of one.

The Yeo-Johnson transformation can be useful in situations where the distribution of the feature is highly skewed or has outliers. By transforming the feature values, the Yeo-Johnson transformation can help to normalize the distribution, making it more suitable for modeling purposes.

The optimal value of the λ parameter can be found by maximizing a likelihood function or by cross-validation. It's important to note that the Yeo-Johnson transformation can only be applied to features with strictly positive values or features that have both positive and negative values.

#### `11. Binarization :` 
This technique converts numerical features to binary features by setting a threshold value. Values above the threshold are set to 1, while values below the threshold are set to 0.

```python
df=pd.read_csv('titanic.csv',usecols=['Age','Fare','Survived'])
df.head()
```

```python
# fillnan
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab 
```

```python
#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)  ## if it falls in a straight line , the we can definitely say that this particular feature is basically a gaussian distrib 
    plt.show()
```

```python
plot_data(df,'Age')
```
