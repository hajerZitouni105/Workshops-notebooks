###**Interpretation**: üìà üí°
We notice that out of the 14 columns containing missing values , only 3 are **numerical** while the other 11 columns are **categorical**.
=> As a result, when dealing with missing values, we will address each type differently.



1.   **For the categorical columns:**

We will replace the missing values with the most frequent value **"mode"** for each respective column. This approach is suitable for categorical data.
2.   **For the  numerical columns:**



Our strategy will depend on "the presence of outliers" within each column. If **outliers are minimal or absent**, we can consider replacing missing values with either the mean or median of the column.

 However, if **outliers are present**, we should exercise caution. In such cases, replacing missing values with the mean might skew the distribution, and it might be more appropriate to use the median or even the mode, depending on the specific characteristics of the data and the context of the analysis.

### **Imputation of categorical columns**


*   **Using SimpleImputer :**



PS: SimpleImputer and fillna() are equivalent

```python
# Create a SimpleImputer with the strategy 'most_frequent'
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
```

```python
df.loc[:,'MasVnrType'] = imputer.fit_transform(df[['MasVnrType']])
```


*   **Using Fillna() :**



```python
# Impute missing values for categorical columns with the most frequent value
categorical_columns = ['BsmtExposure','BsmtFinType2', 'BsmtQual', 'BsmtCond', 'BsmtFinType1','FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond']
for column in categorical_columns:
    df[column].fillna(df[column].mode()[0], inplace=True)
```

###Interpretation: üìà üí°

* We notice that one of the boxplots which belong to 'GarageYrBlt' column does **not contain any outliers**, So the data in that column is relatively well-behaved and normally distributed.

=> For the 'GarageYrBlt' column, using the **mean value** for imputation of missing values is a reasonable approach.

* However, for the columns with outliers which are ['LotFrontage', 'MasVnrArea'], it's essential to handle them carefully to avoid biasing the imputation process. Here are a few strategies you could consider:

1. **Use Median or Mode:** Instead of the mean, you can use the median or the mode value for imputation. Theyn are less sensitive to outliers compared to the mean.
2. **Handle Outliers Separately**: You can identify and handle outliers separately from the imputation process. For example, you could cap or winsorize the outliers to a certain percentile (e.g., replace values above the 95th percentile with the value at the 95th percentile).
3. **Use a Robust Imputation Method:** There are imputation methods specifically designed to handle outliers, such as robust regression imputation or iterative imputation algorithms that are less affected by outliers.

Ultimately, the choice of method depends on the specific characteristics of your data, the nature of the outliers, and the goals of your analysis. It's often a good idea to explore different strategies and assess their impact on your results.






### **Replace the missing values of the 'GarageYrBlt' column**

```python
# Calculate the mean of the column
mean_value = df['GarageYrBlt'].mean()

# Replace missing values with the mean
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(mean_value)
```

### **Replace the missing values of the 'LotFrontage' and 'MasVnrArea' columns**

1. **LotFrontage:**

* Since 'LotFrontage' represents the linear feet of street connected to the property, it's likely correlated with other features such as lot size, neighborhood, or property type.

* One approach could be to impute missing values using **the median** 'LotFrontage' value for properties within the same neighborhood. This can help capture local variations in lot sizes and street frontages.

2. **MasVnrArea:**

* 'MasVnrArea' represents the masonry veneer area in square feet. Since this is a continuous variable, you can consider imputing missing values using **the median or mean** value of the column.

=> For outliers, you may consider a robust imputation method such as Winsorization, where extreme values are replaced with values at a specified percentile (e.g., 95th percentile).

```python
from scipy.stats.mstats import winsorize
```

```python
# Impute missing values for 'LotFrontage' by median within each neighborhood
df['LotFrontage'] = df.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))
```

```python
# Winsorize outliers for 'MasVnrArea'
df['MasVnrArea'] = winsorize(df['MasVnrArea'], limits=[0.1, 0.1])
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[ 'MasVnrArea'], orient='h')
plt.xlabel('Value')
plt.ylabel('MasVnrArea')
plt.title('Boxplots of Numerical Columns')
```

```python
# Impute missing values for 'MasVnrArea' by median of the column
df['MasVnrArea'] = df['MasVnrArea'].fillna(df['MasVnrArea'].mean())
```
